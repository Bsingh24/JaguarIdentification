```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('Data/train.csv')
```

```python
grouping = df.groupby('ground_truth').agg('count').sort_values('filename')
```

```python
index_train = set()
index_valid = set()
index_test = set()

for name, num in zip(grouping.index, grouping.filename):
    index_list = df[df['ground_truth'] == name].index
    train_list = list(np.random.choice(index_list, size=round(num*0.7), replace=False))
    index_train.update(train_list)
    index_list = [item for item in index_list if item not in train_list]
    valid_list = list(np.random.choice(index_list, size=round(num*0.2), replace=False))
    index_valid.update(valid_list)
    index_list = [item for item in index_list if item not in valid_list]
    test_list = list(np.random.choice(index_list, size=round(num*0.1), replace=False))
    index_test.update(test_list)

    print(f'Label: {name}\n Sample Size: {num}\n Training Size: {len(train_list)}\n Validation Size: {len(valid_list)}\n Test Size: {len(test_list)}')
```

```python
categories = list(grouping.index)
enc = OrdinalEncoder(categories=[categories], dtype=np.int32)
df['encoded_ground_truth'] = enc.fit_transform(df[['ground_truth']])
```

```python
final_df = df.drop('ground_truth', axis=1).copy()
```

```python
train_df = final_df.iloc[list(index_train)]
valid_df = final_df.iloc[list(index_valid)]
test_df = final_df.iloc[list(index_test)]
```

```python
train_df.to_csv('Data/train_data.csv', index=False)
valid_df.to_csv('Data/valid_data.csv', index=False)
test_df.to_csv('Data/test_data.csv', index=False)
```
